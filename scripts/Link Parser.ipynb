```python
import config
import pandas as pd
import io
import requests
import numpy as np
```

Source 'https://fantasyoverlord.com/FPL/History?fbclid=IwAR0gWIjMOGRwXomk4aNxfRslfJlpzaaEtm2YAs6wXMEWnLIEq3eY8Ks_xUo'

```python
combined_df = pd.DataFrame()
for i, url in enumerate(config.file_links_2016):
    print(url)
    print(f'{i+1}/{len(config.file_links_2016)}')
    s=requests.get(url).content
    df=pd.read_csv(io.StringIO(s.decode('utf-8')))
    df['url'] = url
    combined_df = combined_df.append(df)
```

```python
combined_df.columns
```

```python
combined_df['season'] =combined_df.url.str.split('-|=').apply(lambda x: x[1]).str.replace('FPL','').astype(int) + 2000
combined_df['gw_gameweek'] =combined_df.url.str.split('-|=').apply(lambda x: x[2]).str.replace('GW','').astype(int)
```

```python
combined_df.to_csv('../data/2016_season_raw.csv')
```
