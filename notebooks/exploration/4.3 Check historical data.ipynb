### Correction

```python
full_season_data_1718[
    (full_season_data_1718['name'] == 'virgil_van dijk')
]['team_name'].value_counts()
```

```python
full_season_data_1718['team_name'] = np.where(
    (full_season_data_1718['name'] == 'virgil_van dijk') &
    (full_season_data_1718['kickoff_time'] < '2018-01-01'),
    'Southampton',
    full_season_data_1718['team_name']
)
```

```python
full_season_data_1718[
    (full_season_data_1718['name'] == 'virgil_van dijk')
]['team_name'].value_counts()
```
