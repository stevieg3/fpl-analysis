```python
del calculated_pfi['total_points_plus1_gw']
```

```python
plot_df = pd.DataFrame(calculated_pfi, index=[0]).transpose().reset_index()
plot_df.rename(columns={'index': 'feature', 0: 'PFI'}, inplace=True)
plot_df.sort_values('PFI', ascending=False, inplace=True)
plot_df.reset_index(drop=True, inplace=True)
plot_df.head(10)
```

```python
plt.figure(figsize=(10,12))
plt.xlim(left=1)
plt.axvline(x=1, color='black')
plt.barh(plot_df['feature'], plot_df['PFI'])
plt.gca().invert_yaxis()
```
